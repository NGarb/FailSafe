```python
from tpot import TPOTRegressor
import pandas as pd 
import numpy as np
```

```python
X_train = pd.read_csv('D:/Projects/Prediction/Techniques Practice/Trees/XGBoost/Suicides/temp data/X_train.csv').drop('Unnamed: 0',axis=1)
y_train = pd.read_csv('D:/Projects/Prediction/Techniques Practice/Trees/XGBoost/Suicides/temp data/y_train.csv').drop('Unnamed: 0',axis=1)
X_val = pd.read_csv('D:/Projects/Prediction/Techniques Practice/Trees/XGBoost/Suicides/temp data/X_val.csv').drop('Unnamed: 0',axis=1)
y_val = pd.read_csv('D:/Projects/Prediction/Techniques Practice/Trees/XGBoost/Suicides/temp data/y_val.csv').drop('Unnamed: 0',axis=1)
X_test = pd.read_csv('D:/Projects/Prediction/Techniques Practice/Trees/XGBoost/Suicides/temp data/X_test.csv').drop('Unnamed: 0',axis=1)
y_test = pd.read_csv('D:/Projects/Prediction/Techniques Practice/Trees/XGBoost/Suicides/temp data/y_test.csv').drop('Unnamed: 0',axis=1)
```

```python
tpot = TPOTRegressor(verbosity=2, max_time_mins=2, max_eval_time_mins=0.04, population_size=200)
tpot.fit(X_train, y_train)
# To get the score 
tpot.score(X_test, y_test)
```

```python
# Export the pipeline as a python file
tpot.export('tpot_suicides_pipeline.py')
```

```python
#pipeline found a decision tree to fit best! WHAT! almost as good as xgboost?
from sklearn.tree import DecisionTreeRegressor
exported_pipeline = DecisionTreeRegressor(max_depth=8, min_samples_leaf=19, min_samples_split=13)

exported_pipeline.fit(X_train, y_train)
results = exported_pipeline.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error as mse
np.sqrt(mse(results,y_test))
```

```python
#okay well like not close to xgb which was 0.04 but not bad - much better than 0.6

exported_pipeline.feature_importances_
exported_pipeline.tree_
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(exported_pipeline, max_depth = 3,
                   feature_names=X_train.columns,  
                   filled=True)
```

```python
import graphviz
# DOT data
dot_data = tree.export_graphviz(exported_pipeline, out_file=None, 
                                feature_names=X_train.columns,  
                                filled=True)

# Draw graph
graph = graphviz.Source(dot_data, format="png") 
graph
```
