```python
# to understand the data a bit better, we will construct a well-fitting pruned decision tree
X_train, X_test, y_train, y_test = train_test_split(bike_hour_df.drop('cnt',axis=1), bike_hour_df.cnt, test_size=0.2)


dt_reg = DecisionTreeRegressor()
dt_reg.fit(X_train, y_train)
```

```python
y_pred = dt_reg.predict(X_test)
```

```python
results_df = pd.DataFrame(columns = ['model','rmse','r2','mape'])
row_obj = reg_metrics.calc_results(y_pred, y_test, 'Decision Tree')
row = row_obj.calc_results_row()
results_df = results_df.append(row, ignore_index=True)
results_df
```

```python
# Decision tree actually models the data really well. But we are mostly doing this to get a sense in exploration
```

```python
# how big is the tree?

print('This tree deep:',dt_reg.get_depth())
```

```python
# prune tree to get something to work with. 
path = dt_reg.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas
alpha_vals=[]
for ccp_alpha in ccp_alphas:
    dt_reg = DecisionTreeRegressor(ccp_alpha=ccp_alpha)
    scores = cross_val_score(dt_reg, X_train, y_train, cv = 5)
    alpha_vals.append([ccp_alpha, np.mean(scores), np.std(scores)])

alpha_results_df = pd.DataFrame(alpha_vals, columns = ['alpha','mean_accuracy','std'])

alpha_results_df.plot(x = 'alpha', y = 'mean_accuracy', yerr = 'std', marker = 'o', linestyle='--', figsize=(15, 6))
```
